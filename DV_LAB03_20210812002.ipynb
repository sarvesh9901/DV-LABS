```python
# Import necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns

# Load dataset
titanic = sns.load_dataset("titanic")

# Plot histogram
sns.histplot(data = titanic, x = "fare",y= "class",discrete=True)
```

```python
sns.histplot(data = titanic, x = "fare",discrete=True)
```

```python
sns.histplot(data = titanic, x = "fare",y="who",discrete=True) # price distribution as per passengers
```

```python
sns.boxplot(data=titanic, x="age", y="sex", hue="alive") #box plot for sex age and alive / Dead
```
